```python
soup= BeautifulSoup(webpage.content, "html.parser")
```

```python
print(soup.prettify())  #prettify() enable us to view how tags are nested in document
```

```python
names=soup.find_all('div',class_="_2WkVRV")
```

```python
names
```

```python
Bag_names=[]
for i in range(0,len(names)):
  Bag_names.append(names[i].get_text())
```

```python
Bag_names
```

```python
Price=soup.find_all('div',class_="_3I9_wc")
```

```python
Price
```

```python
price_list=[]
for i in range(0,len(Price)):
  price_list.append(Price[i].get_text())
price_list
```

```python
import pandas as pd
df=pd.DataFrame()
```

```python
df['Names']=Bag_names
df['Price']=price_list
```

```python
df
```

```python
df.to_csv('Bags.csv',index=False)
```
