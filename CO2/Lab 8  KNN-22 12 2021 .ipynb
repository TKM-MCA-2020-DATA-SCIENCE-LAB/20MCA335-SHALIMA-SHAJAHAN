```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('/content/sample_data/sonar_csv.csv')
```

```python
from sklearn.preprocessing import StandardScaler
```

scaler=StandardScaler

```python
scaler=StandardScaler()
```

```python
scaler.fit(dataset.drop('Class',axis=1))
```

```python
dataset.info()
```

```python
scaled_features=scaler.transform(dataset.drop('Class',axis=1))
```

```python
scaled_features
```

```python
new_dataset=pd.DataFrame(scaled_features,columns=dataset.columns[:-1])
```

```python
new_dataset
```

```python
from sklearn.model_selection import train_test_split
X=new_dataset
Y=dataset['Class']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,Y_train)
KNeighborsClassifier(n_neighbors=1)
```

```python
ped=knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(Y_test,ped))
print(classification_report(Y_test,ped))
```

```python
#analyzing better k value through iterations

error_rate=[]

for i in range(1,40):
  knn=KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train,Y_train)
  ped_i=knn.predict(X_test)
  error_rate.append(np.mean(ped_i !=Y_test))
  
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue',linestyle='dashed',marker='s',markerfacecolor='orange',markersize=10)
plt.title('Error rate vs values')
plt.xlabel('K')
plt.ylabel('Error rate')
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,Y_train)
ped=knn.predict(X_test)
print(confusion_matrix(Y_test,ped))
print(classification_report(Y_test,ped))
```
