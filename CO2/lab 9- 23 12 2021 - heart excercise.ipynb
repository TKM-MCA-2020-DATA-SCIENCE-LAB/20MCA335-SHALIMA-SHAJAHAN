```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset= pd.read_csv("/content/sample_data/Heart_Disease_Prediction.csv")
dataset
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(dataset.drop('Heart Disease',axis=1))
```

```python
dataset.columns[:-1]
```

```python
scaled_features= scaler.transform(dataset.drop('Heart Disease',axis=1))
scaled_features
```

```python
new_dataset=pd.DataFrame(scaled_features,columns=dataset.columns[:-1])
new_dataset
```

```python
from  sklearn.model_selection import train_test_split
X=new_dataset
Y=dataset['Heart Disease']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,CategoricalNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
disease=[]
```

```python
classifier=GaussianNB()
classifier.fit(X_train,Y_train)
Y_pred=classifier.predict(X_test)
disease.append(accuracy_score(Y_test,Y_pred))
cm= confusion_matrix(Y_test,Y_pred)
```
