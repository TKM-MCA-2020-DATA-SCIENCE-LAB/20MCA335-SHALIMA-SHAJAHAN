```python
import matplotlib.pyplot as plt

import numpy as np
x=np.linspace(0,5,11)
y=x**2
x
```

```python
y
```

basic matplotlib commands

```python
plt.plot(x,y,'r')
plt.xlabel('X Axis')
plt.ylabel('y Axis')
plt.title('String title')
plt.show()
```

create multiplot on same function

```python
#plt.subplot(nrows,ncol,plot_number)
plt.subplot(1,3,1)
plt.plot(x,y,'r--')
plt.subplot(1,3,2)
plt.plot(x,y,'g*-')
plt.subplot(1,3,3)
plt.plot(x,y,'r--')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(x,y,'g*-')
```

```python
plt.subplot(1,4,1)
plt.plot(x,y,'r--')
plt.subplot(1,4,2)
plt.plot(x,y,'g*-')
plt.subplot(1,4,3)
plt.plot(x,y,'r--')
plt.subplot(1,4,4)
plt.plot(x,y,'g*-')
```

matplot using object oriented 

```python
#create figure(empty canvas)
fig=plt.figure()

#add set of axes to figure
axes = fig.add_axes([0.1,0.1,0.8,0.8])    #(range 0 to 1)
axes.plot(x,y,'b')
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('title')
```

```python
fig=plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])    #main
axes2 =fig.add_axes([.2,.5,.4,.4])

axes1.plot(x,y,'b')
axes1.set_xlabel('X1')
axes1.set_ylabel('Y1')
axes1.set_title('title')

axes2.plot(y,x,'r--')
axes2.set_xlabel('X2')
axes2.set_ylabel('Y2')
axes2.set_title('title2')
```

LEGENDS,LABELS AND TITLES


```python
fig, ax = plt.subplots()

ax.plot(x, x**2, label="y = x**2")
ax.plot(x, x**3, label="y = x**3")
ax.legend(loc=2); # upper left corner
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title');
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

ax.plot(x, x**2, label="y = x**2")
ax.plot(x, x**3, label="y = x**3")
ax.legend(loc=2);
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title');
```

SETTING COLORS,LINEWIDTH AND LINETYPES

```python
import matplotlib.pyplot as plt
fig, ax = plt. subplots()
ax.plot(x, x+1, color="blue", alpha=0.5)
ax.plot(x, x+2, color="#8B008B") 
ax.plot(x, x+3, color="#FF8C00")
```

line and marker style

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
ax.plot(x, x+4, color="green", lw=3, ls='-.')
ax.plot(x, x+6, color="yellow", lw=9, ls=':')

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')

# marker size and color
ax.plot(x, x+11, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+13, color="purple", lw=2, ls='-', marker='s', markersize=7, markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```
