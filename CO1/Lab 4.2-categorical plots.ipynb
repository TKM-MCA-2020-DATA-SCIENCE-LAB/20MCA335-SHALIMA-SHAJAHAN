CATEGORICAL PLOTS

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

BARPLOT

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

COUNTPLOT

```python
sns.countplot(x='sex',data=tips)
```

BOXPLOT:
Used for comparison between variables

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

to set horizontally do

```python
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="dark")
```

VIOLINPLOT
the violin plot features a kernel density estimation of the underlying distribution.

```python
sns.violinplot(x="day", y="total_bill",hue='sex', data=tips,palette='rainbow')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=True,palette='colorblind')
```

STRIPPLOT : draw a scatterplot where one variable is categorical

SWARMPLOT: similar to stripplot but points are adjusted(no overlap)

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='rainbow',split=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```
